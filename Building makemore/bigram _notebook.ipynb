Now we need to find how often these words appear in our data and to do that we need to maintain some kind of dictionar

```python
b={}
for w in words[:3]:
    #Creating special character
    chs= ['<S>'] + list(w) + ['</E>']
    for ch1, ch2 in zip(chs,chs[1:]):
        bigram=(ch1,ch2)
        b[bigram]=b.get(bigram,0)+1
        print(ch1,ch2)
```

#Now this table needs to be filled with integer , hence we need to have lookup table whihc convert character into integer
- Counting bigram in a 2D torch tensor

```python
chars=sorted(list(set(''.join(words))))
stoi= {s:i for i,s in enumerate(chars)}
```

```python
stoi
```

```python
# Lest add our special character 
stoi['<S>']=26
stoi['<E>']=27
```

```python
N= torch.zeros( (28, 28) ,  dtype=torch.int32)
```

```python

for w in words:
    #Creating special character
    chs= ['<S>'] + list(w) + ['<E>']
    for ch1, ch2 in zip(chs,chs[1:]):
        ix1=stoi[ch1]
        ix2=stoi[ch2]
        N[ix1,ix2]+=1
    
```
